```python
df_s = pd.read_excel('seoulPopulation.xls', header = 2, usecols='B,D,G,J,N')
df_s
```

```python
df_s.rename(columns = {df_s.columns[0]:'구별',
                       df_s.columns[1]:'인구수', df_s.columns[2]:'한국인', df_s.columns[3]:'외국인',
                       df_s.columns[4]:'고령자'}, inplace = True)

df_s
```

```python
df_s.drop([0], inplace = True)
df_s
```

```python
df_s['구별'].unique()
```

```python
df_s.drop([26], inplace = True)
df_s
```

```python
df_s['외국인비율'] = df_s['외국인']/df_s['인구수'] * 100
df_s['고령자비율'] = df_s['고령자']/df_s['인구수'] * 100
df_s.head()
```

```python
df_s.sort_values(by = '인구수', ascending= False)
```

```python
df = pd.merge(df_c, df_s, on = '구별')
df
```

```python
del df['2013년도 이전']
del df['2014년']
del df['2015년']
del df['2016년']

df
```

```python
df.set_index('구별', inplace = True)
df
```

```python
df.sort_values(by = '소계', ascending= False)
```

```python
df.sort_values(by = '인구수', ascending= False)
```

```python
import matplotlib.pyplot as plt
df['소계'].plot(kind = 'barh', grid = True, figsize = (10,10))
plt.show()
```

```python
df['소계'].sort_values().plot(kind = 'barh', grid = True, figsize = (10,10))
plt.show()
```

```python
df['CCTV비율'] = df['소계']/ df['인구수'] * 100
```

```python
df['CCTV비율'].sort_values().plot(kind = 'barh', grid = True, figsize = (10,10))
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.scatter(df['인구수'],df['소계'], s = 50)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
import numpy as np
np.set_printoptions(suppress = True)

fp1 = np.polyfit(df['인구수'], df['소계'], 1)

fp1
```

```python
f1 = np.poly1d(fp1)
print(f1,'\n')

fx = np.linspace(100000, 700000, 100)
print(fx)
```

```python
plt.figure(figsize = (10,10))
plt.scatter(df['인구수'], df['소계'], s=50)
plt.plot(fx,f1(fx), ls = 'dashed', lw = 3, color = 'g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(df['인구수'], df['소계'], 1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000,700000,100)

df['오차'] = np.abs(df['소계'] - f1(df['인구수']))

df_sort = df.sort_values(by = '오차', ascending= False)

df_sort
```

```python
plt.figure(figsize = (14,10))
plt.scatter(df['인구수'], df['소계'], c = df['오차'], s = 50)
plt.plot(fx, f1(fx), ls = 'dashed', lw = 3, color = 'g')

for n in range(10):
  plt.text(df_sort['인구수'][n]*1.02,
           df_sort['소계'][n]*0.98,
           df_sort.index[n])
plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```
